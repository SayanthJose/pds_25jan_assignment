Identifying a Pattern
To explore an interesting pattern in the dataset, I examined the relationship between weather conditions and new bike‑share usage. Since outdoor activity is strongly influenced by temperature, I focused on whether warmer weather corresponds to higher bike‑share counts. This relationship is meaningful because it can help explain how environmental factors affect transport behaviour.

The two datasets (part2a.csv and part2b.csv) were first inspected using .info(), .describe(), and missing‑value summaries. Both datasets contained a timestamp column, which allowed them to be merged using an inner join.
The timestamp column was then converted to a proper datetime format which ensured correct interpretation of the DD/MM/YYYY structure.
The merged dataset included weather variables (temperature, feels‑like temperature, humidity, wind speed) alongside the number of new bike‑share rentals.

Observed Trend
I created several visualisations to explore the data:

A histogram of new_bike_shares showed a right‑skewed distribution, indicating that high‑usage periods are less common.

Boxplots of the weather variables revealed their typical ranges and highlighted outliers.

Scatterplots comparing bike‑share usage with temperature, feels‑like temperature, humidity, and wind speed provided insight into potential relationships.

The most noticeable pattern appeared in the scatterplot of temperature vs. new bike‑share usage, which showed a clear positive upward trend. As temperature increases, the number of new bike‑share rentals tends to rise.

```python
# -----------------------------
# Load datasets
# -----------------------------
df_a = pd.read_csv("part2a.csv")
df_b = pd.read_csv("part2b.csv")

print("Dataset A Overview:")
display(df_a.info(), df_a.describe(), df_a.isna().sum())

print("\nDataset B Overview:")
display(df_b.info(), df_b.describe(), df_b.isna().sum())

# -----------------------------
# Merge datasets on timestamp
# -----------------------------
df = pd.merge(df_a, df_b, on="timestamp", how="inner")

print("\nMerged Dataset Overview:")
display(df.info(), df.describe(), df.isna().sum())

# Convert timestamp to datetime (DD/MM/YYYY HH:MM)
df["timestamp"] = pd.to_datetime(
    df["timestamp"],
    format="%d/%m/%Y %H:%M"
)

# -----------------------------
# Visualisations
# -----------------------------

# Histogram of new bike shares
plt.figure(figsize=(8,4))
sns.histplot(df["new_bike_shares"], kde=True)
plt.title("Distribution of New Bike Shares")
plt.xlabel("New Bike Shares")
plt.ylabel("Frequency")
plt.show()

# Boxplots for weather variables
plt.figure(figsize=(10,5))
df[["temperature", "feels_like_temp", "humidity", "wind_speed"]].boxplot()
plt.title("Boxplots of Weather Variables")
plt.ylabel("Value")
plt.show()

# Scatterplots: bike shares vs weather variables
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.scatterplot(ax=axes[0,0], data=df, x="temperature", y="new_bike_shares")
sns.scatterplot(ax=axes[0,1], data=df, x="feels_like_temp", y="new_bike_shares")
sns.scatterplot(ax=axes[1,0], data=df, x="humidity", y="new_bike_shares")
sns.scatterplot(ax=axes[1,1], data=df, x="wind_speed", y="new_bike_shares")

axes[0,0].set_title("Bike Shares vs Temperature")
axes[0,1].set_title("Bike Shares vs Feels-Like Temperature")
axes[1,0].set_title("Bike Shares vs Humidity")
axes[1,1].set_title("Bike Shares vs Wind Speed")

plt.tight_layout()
plt.show()
```

## Q2.2: Detect and Remove Any Outliers in the Data Used for Your Interesting Pattern [10 marks]

Use an appropriate technique to detect and remove any outliers in the data
used for your interesting pattern or trend.
You must include an explanation of the detection method used, how it works,
and any outliers detected.
NOTE: there may not be any detectable outliers using the selected detection

To detect and remove outliers from the data used for my interesting pattern, I applied the Interquartile Range (IQR) method, which is a standard and reliable technique for identifying extreme values in numerical datasets.

The method works by calculating the first quartile (Q1) and third quartile (Q3), then computing the IQR as:

IQR = Q3 - Q1
Any value below

Q1 - 1.5 x IQR
or above

Q3 + 1.5 x IQR
is considered an outlier.

Applying this method to all numerical variables (new_bike_shares, temperature, feels_like_temp, humidity, and wind_speed) resulted in 1006 outliers being detected out of 17,414 rows. After removing these rows, the cleaned dataset contained 16,408 observations. This reduction is expected because bike‑share and weather data naturally contain occasional spikes and extreme values.

Reference:  
GeeksforGeeks (2025). Interquartile Range to Detect Outliers in Data. Available at: https://www.geeksforgeeks.org/machine-learning/interquartile-range-to-detect-outliers-in-data/ (geeksforgeeks.org in Bing) [Accessed 06 May 2026].

```python
# IQR Outlier Detection for all numerical columns

numeric_cols = ["new_bike_shares", "temperature", "feels_like_temp",
                "humidity", "wind_speed"]

iqr_outliers = pd.Series(False, index=df.index)

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    iqr_outliers |= (df[col] < lower) | (df[col] > upper)

print("Number of IQR outliers detected:", iqr_outliers.sum())

# Remove outliers
df_clean = df[~iqr_outliers].copy()

print("Original dataset size:", len(df))
print("Cleaned dataset size:", len(df_clean))
```

## Q2.3: Hypothesis Test [10 marks]

Define a hypothesis using an appropriate hypothesis testing formulation.

Using an appropriate Python library, test this hypothesis.
You must include a detailed explanation of your findings to achieve good marks for this task.

Hypothesis:  
To test whether temperature has a statistically significant relationship with new bike‑share usage.

Null Hypothesis (H₀): Temperature has no linear relationship with new bike shares.

Alternative Hypothesis (H₁): Temperature does have a linear relationship with new bike shares.

Choice of Test:  
The Pearson correlation test is used because both variables (temperature and new bike shares) are continuous, and the test measures the strength and direction of a linear relationship between them.

Results:  
The test produced a correlation coefficient of 0.37 and a p‑value of 0.0.
Since the p‑value is far below 0.05, H₀ is rejected.

Conclusion:  
There is a statistically significant relationship between temperature and new bike‑share usage. As temperature increases, bike‑share activity tends to increase as well.

```python
# Pearson correlation test between temperature and new bike shares

from scipy.stats import pearsonr

# Extract variables from cleaned dataset
x = df_clean["temperature"]
y = df_clean["new_bike_shares"]

# Perform Pearson correlation test
corr, p_value = pearsonr(x, y)

print("Correlation coefficient:", corr)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Result: Significant relationship (reject H0)")
else:
    print("Result: No significant relationship (fail to reject H0)")
```
