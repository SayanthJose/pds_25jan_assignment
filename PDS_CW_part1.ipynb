## Q1.1: Reading in rows of a CSV file. [10 marks]

Write a function `csv_rows(filename, from_row=None, to_row=None)`
to read in the header and specific rows of a CSV file.
The input arguments of the function should be a string specifying a filename,
a starting row number
`from_row` and
a final row number `to_row`.
The row numbers should be either integers or None.
The starting row number can be 1 or higher or None.
Row 1 refers to the first row after the header.
If the starting row number is None, then the default value of 1 should be used.
The ending row number can be 1 or higher or None.
The ending row number refers to the last row to be included.
If the ending row number is None, then all rows starting at the starting row should be read in.
The function should return two outputs:
a list with the column names and list of rows, where each row is a list of values (strings or numbers).

You can use the file `part1.csv` for your tests.

The function csv_rows reads a CSV file using only built‑in Python.
The first line is split into column names, and the remaining lines are treated as data rows.

from_row and to_row allow selecting a specific range of rows (1‑based indexing after the header).
If from_row is None, it defaults to 1.
If to_row is None, all rows from the starting point are included.

Each value is processed with convert_value, which converts numeric strings into integers or floats and keeps missing values (e.g., "-") as strings.

The function returns:

a list of column names

a list of processed data rows

```python
def convert_value(x):
    """
    Converts a string value into int, float, or keeps it as a string.
    Handles empty strings and '-' as non-numeric values.
    """
    x = x.strip()

    # Treat empty or '-' as plain strings
    if x == "" or x == "-":
        return x

    # Check for integer (including negative)
    if x.lstrip("-").isdigit():
        return int(x)

    # Check for float (one decimal point, digits on both sides)
    parts = x.split(".")
    if len(parts) == 2 and parts[0].lstrip("-").isdigit() and parts[1].isdigit():
        return float(x)

    # Otherwise keep as string
    return x


def csv_rows(filename, from_row=None, to_row=None):
    """
    Reads selected rows from a CSV file and returns the header and data rows as lists.

    
    """

    # Default starting row
    if from_row is None:
        from_row = 1

    # Read file manually 
    f = open(filename, "r")
    lines = f.read().splitlines()
    f.close()

    # Extract header
    headers = lines[0].split(",")

    data_rows = []

    # Loop through data rows (line index 1 = row 1)
    for i in range(1, len(lines)):
        row_number = i  # row_number = 1 for first data row

        # Skip rows before from_row
        if row_number < from_row:
            continue

        # Stop if past to_row
        if to_row is not None and row_number > to_row:
            break

        # Split row and convert values
        raw_values = lines[i].split(",")
        row = [convert_value(v) for v in raw_values]
        data_rows.append(row)

    return headers, data_rows
```

```python
header, rows = csv_rows("part1.csv", from_row=2, to_row=5)

print("Header:", header)
print("Rows:")
for r in rows:
    print(r)

```

## Q1.2: Reading in Columns of a CSV file. [10 marks]

Write a function `csv_cols(filename, from_row=None, to_row=None, columns=None)`
that takes as inputs a filename, a starting row, a final row and a list of column names.
The same conventions as for Q1.1 should be used for the starting and final row input parameters.
The column name parameter can either be a list of column names or None, which indicates all columns.
The function should return a dictionary with the selected column names as
keys and a list of values (strings or numbers) for the corresponding column.
Your function should use the function from Q1.1.

You can use the file `part1.csv` for your tests.




The function csv_cols extracts selected columns from a CSV file.
It first calls csv_rows (from Q1.1) to read the header and the chosen row range.
The columns argument may be a list of column names or None; if None, all columns are returned.

A dictionary is then created where each key is a column name and each value is a list of values from that column.
Numeric conversions and missing‑value handling from Q1.1 are preserved.
The final output is a dictionary containing only the requested columns.

```python
def csv_cols(filename, from_row=None, to_row=None, columns=None):
    

    # Use Q1.1 to read the rows
    headers, rows = csv_rows(filename, from_row, to_row)

    # If no specific columns requested, return all
    if columns is None:
        columns = headers

    # Prepare dictionary with empty lists
    col_dict = {col: [] for col in columns}

    # Fill dictionary with values from each row
    for row in rows:
        for i in range(len(headers)):
            col_name = headers[i]
            if col_name in col_dict:
                col_dict[col_name].append(row[i])

    return col_dict
```

```python
result = csv_cols("part1.csv", from_row=2, to_row=5, columns=None)
print(result)

```

## Q1.3: Calculating the minimum, maximum and median of a list of values [10 marks]

Write a function `min_max_median(p)` to calculate
the minimum, maximum and median of a list of values.
The function should return the minimum, maximum and median.




The function min_max_median takes a list of numeric values and returns three summary statistics: the minimum value, the maximum value, and the median.
The minimum and maximum are computed manually using a loop, without using Python’s built‑in min() or max() functions.

To compute the median, a copy of the list is created and sorted.
If the list has an odd number of elements, the median is the middle value.
If the list has an even number of elements, the median is calculated as the average of the two central values.

The function assumes the input list contains only numeric values (integers or floats), which is consistent with the processed output from earlier questions.
It returns a tuple in the form (minimum, maximum, median).

```python
def min_max_median(values):
    """
    Computes the minimum, maximum, and median of a list of numeric values.
    Returns a tuple (min_value, max_value, median_value).
    """

    # Compute minimum and maximum manually
    min_value = values[0]
    max_value = values[0]

    for v in values[1:]:
        if v < min_value:
            min_value = v
        if v > max_value:
            max_value = v

    # Sort a copy for median
    sorted_vals = values[:]   # copy
    sorted_vals.sort()

    n = len(sorted_vals)

    # Odd length → middle element
    if n % 2 == 1:
        median = sorted_vals[n // 2]
    else:
        # Even length → average of two middle elements
        median = (sorted_vals[n // 2 - 1] + sorted_vals[n // 2]) / 2

    return min_value, max_value, median
```

```python
print(min_max_median([1, 2, 3, 4, 5]))       # (1, 5, 3)
print(min_max_median([10, 2, 8, 4]))         # (2, 10, 6.0)
print(min_max_median([7]))                   # (7, 7, 7)
```

## Q1.4: Calculating the Minimum, Maximum and Median of Rows and Columns of a Table [10 marks]

Write a function `table_min_max_median(data, columns=None)` that takes as inputs
a dictionary with column names as keys and a list of values for each key.
The `columns` parameter can be either a list of column names or None, which
means all column names should be used.
The function should return a list of (minimum, maximum, median) tuples for each row,
and a list of (minimum, maximum, median) tuples for the selected columns.
Your function must use the function from Q1.3.

You can use the file `part1.csv` and the function from Q1.2 for your tests.

The function table_min_max_median takes a dictionary where each key is a column name and each value is a list of numeric values. This matches the structure returned by csv_cols in Q1.2.
If columns is None, all columns in the dictionary are used; otherwise, only the specified columns are included.

The function computes two types of summary statistics:

Row statistics
For each row index, the values across the selected columns are collected and passed to min_max_median (from Q1.3).
This produces a list of (min, max, median) tuples, one per row.

Column statistics
For each selected column, the list of values in that column is passed to min_max_median.
This produces a (min, max, median) tuple for each column.

The function returns both results as a pair:

a list of row statistics

a list of column statistics

```python
def table_min_max_median(data, columns=None):
    """
    Compute (min, max, median) for each row and each selected column.
    'data' is a dictionary: {column_name: list_of_values}.
    Returns (row_stats, col_stats).
    """

    # Use all columns if none specified
    if columns is None:
        columns = list(data.keys())

    # Number of rows = length of any selected column
    num_rows = len(data[columns[0]])

    # ---- Row stats ----
    row_stats = []
    for i in range(num_rows):
        row_values = []
        for col in columns:
            v = data[col][i]
            if v != "-":      # ignore missing values
                row_values.append(v)
        row_stats.append(min_max_median(row_values))

    # ---- Column stats ----
    col_stats = []
    for col in columns:
        col_values = []
        for v in data[col]:
            if v != "-":      # ignore missing values
                col_values.append(v)
        col_stats.append(min_max_median(col_values))

    return row_stats, col_stats
```

```python
# Use Q1.2 to read all columns and first 5 rows
data = csv_cols("part1.csv", from_row=1, to_row=5, columns=None)

row_stats, col_stats = table_min_max_median(data)

print("Row statistics:")
for r in row_stats:
    print(r)

print("\nColumn statistics:")
for c in col_stats:
    print(c)
```

Write a function `print_min_max_median(row_stats, col_stats, ...)`
that prints the results in a clear and attractive way.
The input parameters should be a list of (min, max, median) tuples for the rows
and a list of (min, max, median) tuples for the columns
(as generated by Q1.4).
You can add extra parameters for further customisation.

You can use the file `part1.csv` and the functions from Q1.2 and Q1.4 for your tests.


The function print_min_max_median displays the row and column statistics produced in Q1.4 in a clear, readable table format.
It takes two lists of (min, max, median) tuples:

row_stats — one tuple per row

col_stats — one tuple per column

The function prints two tables:

Row statistics table
Each line corresponds to a row and shows the minimum, maximum, and median across the selected columns.

Column statistics table
Each line corresponds to a column and shows the minimum, maximum, and median down that column.

```python
def print_min_max_median(row_stats, col_stats, row_label="Row", col_label="Column"):
    """
    Print row and column (min, max, median) statistics in a clear table format.
    """

   
    print("Row Statistics:")
    print(f"{row_label:<10} {'Min':>12} {'Max':>12} {'Median':>12}")
    print("-" * 50)

    for i, (mn, mx, med) in enumerate(row_stats, start=1):
        print(f"{i:<10} {mn:>12.6f} {mx:>12.6f} {med:>12.6f}")

    print("\nColumn Statistics:")
    print(f"{col_label:<10} {'Min':>12} {'Max':>12} {'Median':>12}")
    print("-" * 50)

    for i, (mn, mx, med) in enumerate(col_stats, start=1):
        print(f"{i:<10} {mn:>12.6f} {mx:>12.6f} {med:>12.6f}")
```
